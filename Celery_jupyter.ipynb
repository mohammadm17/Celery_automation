```python
import feedparser
from datetime import datetime

# List of RSS feed URLs
rss_feeds = [
    'https://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml',
    'https://feeds.bbci.co.uk/news/rss.xml',
    # Add more RSS feed URLs as needed
]

# Dictionary to store article information, where keys are URLs and values are dictionaries containing article information
articles_info = {}

# Iterate over each RSS feed URL
for feed_url in rss_feeds:
    # Parse the RSS feed
    feed = feedparser.parse(feed_url)
    
    # Iterate over each entry (news article) in the feed
    for entry in feed.entries:
        # Extract relevant information
        title = entry.title
        content = entry.get('summary', '')  # Some feeds might provide a summary
        
        # Try to get the publication date from various attributes
        if 'published_parsed' in entry:
            publication_date_tuple = entry.published_parsed
        elif 'updated_parsed' in entry:
            publication_date_tuple = entry.updated_parsed
        else:
            # If publication date cannot be determined, use current date and time
            publication_date_tuple = datetime.now().timetuple()
        
        # Convert the publication date tuple to a human-readable format
        publication_date = datetime(*publication_date_tuple[:6]).strftime('%Y-%m-%d %H:%M:%S')
        
        source_url = entry.link
        
        # Check for duplicate articles
        if source_url not in articles_info:
            # Store article information
            articles_info[source_url] = {
                'title': title,
                'content': content,
                'publication_date': publication_date,
                'source_url': source_url
            }

# Print extracted article information
for article_url, article_info in articles_info.items():
    print("Title:", article_info['title'])
    print("Content:", article_info['content'])
    print("Publication Date:", article_info['publication_date'])
    print("Source URL:", article_info['source_url'])
    print("-" * 50)
```
